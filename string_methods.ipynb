# String Methods

```python
text = " Hello Python World "

print(text.lower())
print(text.upper())
print(text.title())
print(text.capitalize())
print(text.strip())
print(text.replace("Python", "Coding"))
print(text.split())
print(text.find("Python"))
print(text.count("o"))
print(text.startswith(" Hello"))
print(text.endswith("World "))
print("Hello".isalpha())
print("123".isdigit())
```
