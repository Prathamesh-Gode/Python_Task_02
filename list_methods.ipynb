# List Methods

```python
nums = [10, 20, 30]

nums.append(40)
print("After append:", nums)

nums.insert(1, 15)
print("After insert:", nums)

nums.extend([50, 60])
print("After extend:", nums)

nums.remove(20)
print("After remove:", nums)

nums.pop()
print("After pop:", nums)

print("Index of 30:", nums.index(30))
print("Count of 10:", nums.count(10))

nums.sort()
print("Sorted:", nums)

nums.reverse()
print("Reversed:", nums)

nums.clear()
print("Cleared:", nums)
```
