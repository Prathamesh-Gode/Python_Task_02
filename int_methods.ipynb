# Int Method

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)

print("Absolute:", abs(-10))
print("Power function:", pow(2, 3))
print("Round:", round(5.6))
print("Max:", max(10, 20, 30))
print("Min:", min(10, 20, 30))
```
